# Download the data  
获取2023-01-01至2023-3-15日 BTC 四小时级别数据(开盘/收盘/最高价/最低价/交易量)  
每天产生24/4=6条数据。为防止计算MA20过程中前19条数据包含缺失值,多取4天,获取2022-12-28至2023-3-15时间段数据

```python
url = "https://api.binance.com/api/v3/klines"

# 参数
symbol = "BTCUSDT"
interval = "4h"
start_time = int(datetime.datetime(2022,12,28).timestamp() * 1000)
end_time = int(datetime.datetime(2023,3,15).timestamp() * 1000)

params = {
    "symbol": symbol,
    "interval": interval,
    "startTime": start_time,
    "endTime": end_time
}
# 发送请求
response = requests.get(url, params=params)

# 将数据转换为DataFrame
data = response.json()
df = pd.DataFrame(data, columns=['open_time', 'open', 'high', 'low', 'close', 'volume', 'close_time', 'quote_asset_volume', 'num_trades', 'taker_base_vol', 'taker_quote_vol', 'ignored'])
df = df[['open_time','open','close','high','low','volume']]
```

# Assignment 2： 
统计时间段内共有多少次涨破或跌破4小时级别MA20，有多少次有效涨破或跌破

```python
import matplotlib.pyplot as plt
plt.plot(df['close'] )
plt.plot(df['MA_20'])
plt.legend(['close', 'MA_20'])
```

```python
count_below,count_below_eff = 0,0
count_above,count_above_eff = 0,0

df.loc[:,'close_lag1'] = df['close'].shift(1)  # 创建一列记录前一行的'close'值
df.loc[:,'ma20_lag1'] = df['MA_20'].shift(1)   # 创建一列记录前一行的'MA_20'值
condition_up = (df['close'] > df['MA_20']) & (df['close_lag1'] <= df['ma20_lag1'])
condition_down = (df['close'] < df['MA_20']) & (df['close_lag1'] >= df['ma20_lag1'])
df_up = df[condition_up]
df_down = df[condition_down]
df = df.drop(['close_lag1','ma20_lag1'],axis=1)
```

```python
# 涨破 
df_up.head()
```

```python
# 跌破
df_down.head()
```

```python
def ifEffective(times,signal):
    lst = []
    for time in times:
        duration = [time+timedelta(hours=i) for i in range(0,24,4)]
        diff =  [df.loc[j,'close'] - df.loc[j,'MA_20'] for j in duration]
        if signal == 1: #涨破
            res = list(filter(lambda x:x < 0,diff)) #如果close一直不低于ma，则有效
        if signal == -1: #跌破
            res = list(filter(lambda x:x > 0,diff))
        if not res:
            lst.append(time)
    return lst
```

```python
eff_up_time = ifEffective(df_up.index,signal=1)
eff_down_time = ifEffective(df_down.index,signal=-1)
df_eff_up = df.loc[eff_up_time,:]
df_eff_down = df.loc[eff_down_time,:]
```

```python
print(f"涨破MA20的次数:{len(df_up)}")
print(f"跌破MA20的次数:{len(df_down)}")
print(f"有效涨破MA20的次数:{len(df_eff_up)}")
print(f"有效跌破MA20的次数:{len(df_eff_down)}")
```

# Assignment 3
计算每次有效涨/跌破MA20后的最大涨跌幅及收盘涨跌幅

```python
# 有效涨破
df_eff_up['closed_updown(%)'] = round(100* (df_eff_up['close'] - df_eff_up['open']) /df_eff_up['open'],2)
df_eff_up.head()
```

```python
# 有效跌破
df_eff_down['closed_updown(%)'] = round(100* (df_eff_down['close'] - df_eff_down['open']) /df_eff_down['open'],2)
df_eff_down.head()
```

```python
def Maxupdown(df1,df2,times):
    for time in times:
        timelst =[ time+timedelta(hours= i) for i in range(0,24,4)]
        lst = [round(100*(df1.loc[time,'close']-df1.loc[time,'open'])/df1.loc[time,'open'],2) for time in timelst]
        uplist = list(filter(lambda x:x>=0,lst))
        downlist= list(filter(lambda x:x<=0,lst))
        max_up = max(uplist)
        max_down = min(downlist)
        df2.loc[time,'max_up_24h(%)'] = max_up
        df2.loc[time,'max_down_24h(%)'] = max_down
    return df2
        
```

```python
df_eff_down = Maxupdown(df,df_eff_down,list(df_eff_down.index))
df_eff_up =  Maxupdown(df,df_eff_up,list(df_eff_up.index))
```
